```python
def explode(table_name, grain = 'day'):
    min_date = table("b2_club").effective_start_date.min()
    max_date = str(datetime.today().date())
    
    grain_map = {
    "day": "D",
    "week": "W-MON",  # Weeks starting on Monday
    "month": "MS",    # Month start
    }
    
    if grain not in grain_map:
        raise ValueError("Unsupported grain: choose from 'day', 'week', or 'month'")
        
    date_range_df = pd.DataFrame({
        "date": pd.date_range(start=min_date, end=max_date, freq=grain_map[grain])
    })
    date_range_df
```
